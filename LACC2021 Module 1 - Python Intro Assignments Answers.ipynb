## English Word to Pig Latin Translator

* Pig Latin is a language game, where you move the first letter of the word to the end and add "ay." So "Python" becomes "ythonpay."

* Prompt the user with the phrase “Enter a word:”, and then output the word entered by the user translated into Pig Latin in lowercase

* If the user’s entry is empty, non-alphabetical, or has multiple word the output should be “Error: incorrect input”.

```python
def checkIfOneWord(input):
    if len(input.split(" ")) > 1 or not input.strip():        
        raise Exception("Error: incorrect input")
    else :
        return 1

def translateToPigLatin(input):
    try:
        checkIfOneWord(input)
        pig_latin = []
        pig_latin.append(input[1:] + input[0] + 'ay')
        print(pig_latin)
    except Exception as error:
        print(error)
        
user_input = input("Enter your input")
print(user_input)
translateToPigLatin(user_input)
```

## Compute Statistics

* Mean, mode, and medianare three common statistics of a set of numbers.

* Prompt the user with the phrase “Enter a comma separated list of numbers:”

* Output on three separate lines the mean, mode, and median of the entered numbers as “Mean = ...”, “Mode = ...” and “Median = ...”. 

* The user may enter numbers as integers or real.

* Hint : Consider using the function input()instead of raw_input()-look up on the web as to how it works.

```python
def mean(input):
    sum = 0.0
    for x in input[:]:
        sum = sum + x
    return sum/len(input)

def median(input):
    input = sorted(input)
    length = len(input)
    if length%2 == 0:
        return (input[length//2] + input[length//2-1])/2
    else:
        return input[length//2]

def mode(input):
    input = sorted(input)
    dict = {}
    for x in input[:]:
        if x in dict:
            dict[x] = dict[x] + 1
        else:
             dict[x] = 1
    maxValue = 0
    mode = 0
    for key in dict:
        if maxValue < dict[key]:
            maxValue = dict[key]
            mode = key
    return mode

input_numbers = input("Enter a comma separated list of numbers ")
input_numbers = [float(x) for x in input_numbers.split(",")]
print(input_numbers)
print("Mean = ", mean(input_numbers))
print("Median = ",median(input_numbers))
print("Mode = ", mode(input_numbers))
```

## Tic-Tac-Toe

* Start a game of tic-tac-toe and print out the initial 3 by 3 board. 

* Prompt a player to make a move through an input. By default, the first player will always put “X”. 

* After every valid player input, the board is updated and printed. 

* Check that a player’s move is valid otherwise the player has to make another input and an error message of "Invalid move" is printed. 

* Game continues until either the game reaches a stalemate or one player is the winner. 



```python
#Example code taken from https://medium.com/byte-tales/the-classic-tic-tac-toe-game-in-python-3-1427c68b8874

theBoard = {'7': ' ' , '8': ' ' , '9': ' ' ,
            '4': ' ' , '5': ' ' , '6': ' ' ,
            '1': ' ' , '2': ' ' , '3': ' ' }

board_keys = []

for key in theBoard:
    board_keys.append(key)

''' We will have to print the updated board after every move in the game and 
    thus we will make a function in which we'll define the printBoard function
    so that we can easily print the board everytime by calling this function. '''

def printBoard(board):
    print(board['7'] + '|' + board['8'] + '|' + board['9'])
    print('-+-+-')
    print(board['4'] + '|' + board['5'] + '|' + board['6'])
    print('-+-+-')
    print(board['1'] + '|' + board['2'] + '|' + board['3'])

# Now we'll write the game code
player = 'X'
turns = 0

while (turns < 10):
    printBoard(theBoard)
    print("It's your turn," + player + ".Move to which place?")

    move = input()        

    #Check if move is valid
    if( move not in board_keys):
      print("Invalid move, move not recognized")
      continue
    
    #Check if space already occupied
    if(theBoard[move] != ' '):
      print("Invalid move, space already occupied")
      continue

    theBoard[move] = player
    turns += 1


    # Now we will check if player X or O has won,for every move after 5 moves. 
    if turns >= 5:
        if theBoard['7'] == theBoard['8'] == theBoard['9'] != ' ': # across the top
            printBoard(theBoard)
            print("\nGame Over.\n")                
            print(" **** " +player + " won. ****")                
            break
        elif theBoard['4'] == theBoard['5'] == theBoard['6'] != ' ': # across the middle
            printBoard(theBoard)
            print("\nGame Over.\n")                
            print(" **** " +player + " won. ****")
            break
        elif theBoard['1'] == theBoard['2'] == theBoard['3'] != ' ': # across the bottom
            printBoard(theBoard)
            print("\nGame Over.\n")                
            print(" **** " +player + " won. ****")
            break
        elif theBoard['1'] == theBoard['4'] == theBoard['7'] != ' ': # down the left side
            printBoard(theBoard)
            print("\nGame Over.\n")                
            print(" **** " +player + " won. ****")
            break
        elif theBoard['2'] == theBoard['5'] == theBoard['8'] != ' ': # down the middle
            printBoard(theBoard)
            print("\nGame Over.\n")                
            print(" **** " +player + " won. ****")
            break
        elif theBoard['3'] == theBoard['6'] == theBoard['9'] != ' ': # down the right side
            printBoard(theBoard)
            print("\nGame Over.\n")                
            print(" **** " +player + " won. ****")
            break 
        elif theBoard['7'] == theBoard['5'] == theBoard['3'] != ' ': # diagonal
            printBoard(theBoard)
            print("\nGame Over.\n")                
            print(" **** " +player + " won. ****")
            break
        elif theBoard['1'] == theBoard['5'] == theBoard['9'] != ' ': # diagonal
            printBoard(theBoard)
            print("\nGame Over.\n")                
            print(" **** " +player + " won. ****")
            break 

    # If neither X nor O wins and the board is full, we'll declare the result as 'tie'.
    if turns == 9:
        print("\nGame Over.\n")                
        print("It's a Tie!!")

    # Now we have to change the player after every move.
    if player =='X':
        player = 'O'
    else:
        player = 'X'                 
```
